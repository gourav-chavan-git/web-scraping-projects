```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://remoteok.io/api"
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)

data = response.json()

jobs = []
for job in data[1:]:
    title = job.get("position", np.nan)
    company = job.get("company", np.nan)
    location = job.get("location", "World Wide").strip() or np.nan
    Company_Logo = job.get("company_logo", np.nan)
    Tags = ', '.join(job.get("tags"))
    jobs.append([title, company, location, Company_Logo, Tags])

df = pd.DataFrame(jobs, columns=["Job Title", "Company", "Location", "Company Logo", "Tags"])
```

```python
df.to_csv("job_listings.csv", index=False)
```
